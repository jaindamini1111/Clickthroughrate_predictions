#### Lets encode our categorical columns for feeding them into our models.

```python
categorical_columns = ["city", "country"]
```

```python
labelencoder = LabelEncoder()
```

```python
data[categorical_columns] = data[categorical_columns].apply(lambda column: labelencoder.fit_transform(column))
```

```python
y = data["clicked_on_ad"]
```

```python
X = data[["daily_time_spent_on_site","age","area_income","daily_internet_usage","city","male","country"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
y_train.value_counts(normalize=True)
```

#### Since the dataset is small and the target output is binary, lets use logistic regression.

```python
lr = LogisticRegression(random_state=0)
```

```python
lr.fit(X_train, y_train)
```

```python
lr_y_pred = lr.predict(X_test)
```

```python
print("Accuracy of Logistic Regression model is: ", accuracy_score(y_test, lr_y_pred))
```

#### We will use Decision Tree Classifier,  and visualize the tree, to see what is the root node of our tree.

```python
tree = DecisionTreeClassifier()
```

```python
tree.fit(X_train, y_train)
```

```python
tree_pred = tree.predict(X_test)
```

```python
print("Accuracy of tree model is: ", accuracy_score(y_test, tree_pred))
```

#### Random Forest Classifier, comes with an amazing attribute which lets us know the significance of features for our model. Lets dig in!

```python
rf = RandomForestClassifier(random_state=0)
```

```python
rf.fit(X_train, y_train)
```

```python
forest_y_pred = rf.predict(X_test)
```

```python
print("Accuracy of Random Forest Classfier is:",accuracy_score(y_test, forest_y_pred))
```

```python
feature_imp = pd.Series(rf.feature_importances_ ,index = X.columns).sort_values(ascending=False)
```

```python
sns.barplot(x = feature_imp, y = feature_imp.index)
plt.xlabel("Feature Importance Score")
plt.ylabel("Features")
plt.title("Visualizing Important Features")
plt.show()
```
